```python
legit =data[data.Class == 0]
fraud =data[data.Class == 1]
```

```python
ls=legit.sample(n=492)
#we are combining fraud and legit
```

```python
#legit have equal no data 
new_data=pd.concat([ls,fraud],axis=0)
```

```python
new_data.head()
```

```python
x=new_data.drop(columns='Class',axis=1)
y=new_data['Class']
```

```python
#x is without Class column
print(x)
```

```python
print(y)
```

```python
#spliting the dataset into training and testing data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2) 
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#logistic regression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
train_pred=lr.predict(x_train)
acc=accuracy_score(train_pred,y_train)
test_pred=lr.predict(x_test)
acc1=accuracy_score(test_pred,y_test)
```

```python
print("train accuracy :",acc)
```

```python
print("test accuracy :",acc1)
```

```python
#support vectoor machine machine learning algorithm
from sklearn import svm
svm=svm.SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
pred2=svm.predict(x_test)
accuracy_score(y_test,pred2)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
pred3 = gnb.predict(x_test)
accuracy_score(y_test,pred3)
```

```python
final_data= pd.DataFrame({'Models':['LR','SVM','NB'],
                          'ACC':[accuracy_score(y_test,test_pred),
                               accuracy_score(y_test,pred2),
                               accuracy_score(y_test,pred3),
                               ]})
```

```python
import seaborn as sns
sns.barplot(x=final_data['Models'], y=final_data['ACC'] , width =0.2)
```
